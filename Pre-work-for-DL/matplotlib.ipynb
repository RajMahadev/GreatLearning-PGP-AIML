# Plotting with Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt


X = np.linspace(-10 , 10, 50)
y = np.cos(X)

plt.plot(X, y)

plt.title("Cos curve through matplotlib")
plt.xlabel("X axis")
plt.ylabel("Cos(X)")

```

```python
# First plot

plt.figure(1)
y = np.cos(X)
plt.plot(X, y)

plt.title("Figure 1   Cos Curve")
plt.xlabel("X_axis")
plt.ylabel("y_label")
```

```python
# Second plot

plt.figure(2)
y = np.sin(X)
plt.plot(X, y)

plt.title("Figure 2   Sine Curve")
plt.xlabel("X_axis")
plt.ylabel("y_label")
```

```python
# Using Subplots


y1 = np.cos(X)

figure_obj = plt.figure(figsize = (8,6))


ax1 = figure_obj.add_subplot(3,2,1)
ax1.plot(X,y1)
ax1.set_ylabel("cos X")

y2 = np.sin(X)
ax2 = figure_obj.add_subplot(3,2,2)
ax2.plot(X, y2)
ax2.set_ylabel("sin X")

y3 =  np.tan(X)
ax3 = figure_obj.add_subplot(3,2,3)
ax3.plot(X, y3 )
ax3.set_ylabel("tan X")
        
y4 = np.tanh(X)
ax4 = figure_obj.add_subplot(3,2,6)
ax4.plot(X, y4)
ax4.set_ylabel("tanh X")
```

```python
# Using Subplots with color


y1 = np.cos(X)

figure_obj = plt.figure(figsize = (8,6))


ax1 = figure_obj.add_subplot(3,2,1)
ax1.plot(X,y1 , color = 'green')
ax1.set_ylabel("cos X")

y2 = np.sin(X)
ax2 = figure_obj.add_subplot(3,2,2)
ax2.plot(X, y2, color = 'brown')
ax2.set_ylabel("sin X")

y3 =  np.tan(X)
ax3 = figure_obj.add_subplot(3,2,3)
ax3.plot(X, y3 , color = 'blue')
ax3.set_ylabel("tan X")
        
y4 = np.tanh(X)
ax4 = figure_obj.add_subplot(3,2,6)
ax4.plot(X, y4 , color = 'red')
ax4.set_ylabel("tanh X")
```

```python
# Using Subplots with color and marker


y1 = np.cos(X)

figure_obj = plt.figure(figsize = (8,6))


ax1 = figure_obj.add_subplot(3,2,1)
ax1.plot(X,y1 , color = 'green' , marker = "_")
ax1.set_ylabel("cos X")

y2 = np.sin(X)
ax2 = figure_obj.add_subplot(3,2,2)
ax2.plot(X, y2, color = 'brown', marker = "o")
ax2.set_ylabel("sin X")

y3 =  np.tan(X)
ax3 = figure_obj.add_subplot(3,2,3)
ax3.plot(X, y3 , color = 'blue' , marker = "*")
ax3.set_ylabel("tan X")
        
y4 = np.tanh(X)
ax4 = figure_obj.add_subplot(3,2,6)
ax4.plot(X, y4 , color = 'red' , marker = "^")
ax4.set_ylabel("tanh X")
```

```python
# Using Subplots with color and marker and legend


y1 = np.cos(X)

figure_obj = plt.figure(figsize = (8,6))


ax1 = figure_obj.add_subplot(3,2,1)
ax1.plot(X,y1 , color = 'green' , marker = "_" , label = 'cos')
ax1.legend(loc='best')
ax1.set_ylabel("cos X")

y2 = np.sin(X)
ax2 = figure_obj.add_subplot(3,2,2)
ax2.plot(X, y2, color = 'brown', marker = "o" , label = 'sin')
ax2.legend(loc='best')
ax2.set_ylabel("sin X")

y3 =  np.tan(X)
ax3 = figure_obj.add_subplot(3,2,3)
ax3.plot(X, y3 , color = 'blue' , marker = "*" , label = 'tan')
ax3.legend(loc='best')
ax3.set_ylabel("tan X")
        
y4 = np.tanh(X)
ax4 = figure_obj.add_subplot(3,2,6) 
ax4.plot(X, y4 , color = 'red' , marker = "^" , label = 'tanh')
ax4.legend(loc='best')
ax4.set_ylabel("tanh X")

```
